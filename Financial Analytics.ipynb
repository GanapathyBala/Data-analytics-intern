```python
# Impute missingvalues with the mean of the column
print("\nImpute missingvalues with the mean of the column")
print("\nmissing values are removed then impute")
df['Sales Qtr - Crore'].fillna(df['Sales Qtr - Crore'].mean(),inplace=True)
print(df)
```

```python
#Remove outliers from column Mar Cap - Crore using z-score
print("\nRemove outliers in Mar Cap - Crore using z-score")
from scipy import stats
z_scores=stats.zscore(df['Mar Cap - Crore'])
df=df[(z_scores<3)] # keep rows with z-score<3

#Remove outliers from column Sales Qtr - Crore using z-score
print("\nRemove outliers in Sales Qtr - Crore using z-score")
from scipy import stats
z_scores=stats.zscore(df['Sales Qtr - Crore'])
df=df[(z_scores<3)] # keep rows with z-score<3
print(df)




```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Assuming you have already loaded your dataset into a DataFrame named df
# and it contains columns: 'Name', 'Mar Cap - Crore', and 'Sales Qtr - Crore'

# Encode the categorical column
categorical_column = 'Name'
label_encoder = LabelEncoder()
df[categorical_column + '_encoded'] = label_encoder.fit_transform(df[categorical_column])

# Drop the original categorical column from the DataFrame
df.drop(columns=[categorical_column], inplace=True)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
